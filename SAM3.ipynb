```python
from gradio_client import Client, handle_file

client = Client("merve/SAM3-video-segmentation")

# 1. Initialiser la session TEXT avec ta vidéo
frame_slider_info, preview_img, md1, md2 = client.predict(
    video={"video": handle_file("foot.mp4")},
    api_name="/_on_video_change_text",
)
print("Video loaded")

# 2. Appliquer le prompt texte "ball" sur la frame 0
#    (ou une autre frame si le ballon est plus visible)
preview_with_ball, md3, md4 = client.predict(
    frame_idx=0,
    text_prompt="ball",
    api_name="/on_text_prompt",
)
print("Text prompt applied")

# 3. Propager les masques sur la vidéo
#    (ici on suppose que pour l’onglet texte c’est /propagate_masks_1,
#     si ça plante, essaie /propagate_masks à la place)
try:
    status, frame_idx = client.predict(
        api_name="/propagate_masks_1",
    )
    print("Propagation done:", status, "last frame:", frame_idx)
except Exception as e:
    print("propagate_masks_1 failed, trying propagate_masks:", e)
    status, frame_idx = client.predict(
        api_name="/propagate_masks",
    )
    print("Propagation done:", status, "last frame:", frame_idx)

# 4. Rendre la vidéo avec les masques
#    (même logique : /_render_video_1 pour l’onglet texte en général)
try:
    rendered = client.predict(
        api_name="/_render_video_1",
    )
except Exception as e:
    print("_render_video_1 failed, trying _render_video:", e)
    rendered = client.predict(
        api_name="/_render_video",
    )

print("Rendered video info:", rendered)
# rendered est de la forme {'video': filepath, 'subtitles': None}
```
